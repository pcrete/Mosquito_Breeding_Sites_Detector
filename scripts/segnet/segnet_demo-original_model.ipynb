# SegNet Original Model
Weights can be downloaded from this link: 
http://mi.eng.cam.ac.uk/~agk34/resources/SegNet/segnet_weights_driving_webdemo.caffemodel

The model should be located in: 
breeding_site_detection/scripts/segnet/Models/

```python
import numpy as np
import os.path
import scipy
import argparse
import scipy.io as sio
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.colors as colors
import matplotlib.cm as cmx
import matplotlib.pyplot as plt
import cv2
import sys
import caffe
import scipy.misc
from PIL import Image
from IPython.display import Image as IPython_display
from tqdm import tqdm
import time
%matplotlib inline

caffe.set_mode_cpu()

model = 'Models/segnet_model_driving_webdemo.prototxt'
weights = 'Models/segnet_weights_driving_webdemo.caffemodel'

net = caffe.Net(model,weights,caffe.TEST)

input_shape = net.blobs['data'].data.shape
output_shape = net.blobs['argmax'].data.shape

label_colours = cv2.imread('Models/camvid12.png').astype(np.uint8)

def process(img_path):
    start = time.time()
    frame = cv2.imread(os.path.join('image',img_path))
    frame = cv2.resize(frame, (input_shape[3],input_shape[2]))

    input_image = frame.transpose((2,0,1))
    input_image = np.asarray([input_image])

    out = net.forward_all(data=input_image)
    end = time.time()

    print 'Executed SegNet in', str(end - start)[:-8], 'seconds'

    segmentation_ind = np.squeeze(net.blobs['argmax'].data)
    segmentation_ind_3ch = np.resize(segmentation_ind,(3,input_shape[2],input_shape[3]))
    segmentation_ind_3ch = segmentation_ind_3ch.transpose(1,2,0).astype(np.uint8)
    segmentation_rgb = np.zeros(segmentation_ind_3ch.shape, dtype=np.uint8)

    cv2.LUT(segmentation_ind_3ch, label_colours, segmentation_rgb)
    segmentation_rgb = segmentation_rgb.astype(float)/255
    
    image = frame[...,::-1]
    segmentation_rgb = segmentation_rgb[...,::-1]
    
    scipy.misc.imsave('image/output/'+img_path[:-4]+'_original.jpg', image)
    scipy.misc.imsave('image/output/'+img_path[:-4]+'_segmented.jpg', segmentation_rgb)
    
    bg = Image.open('image/output/'+img_path[:-4]+'_original.jpg')
    fg = Image.open('image/output/'+img_path[:-4]+'_segmented.jpg')
    overlay = Image.blend(bg, fg, 0.5)
    overlay.save('image/output/'+img_path[:-4]+'_overlay.jpg')

    fig = plt.figure(figsize=(19, 10))

    fig.add_subplot(1,2,1)
    plt.imshow(image)

    fig.add_subplot(1,2,2)
    plt.imshow(segmentation_rgb)
    plt.show()
```

```python
process('1.jpg')
IPython_display('image/color_tags.png')
```
